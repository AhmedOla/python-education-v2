```
 
class Animal():
 
    def __init__(self,name,legs,color):
 
        self.name1 = name
 
        self.legs = legs
        
        self.color = color
 
    def print_features(self):
        print("My name is " + self.name1 + ", I have " + str(self.legs) + " legs" +  " my color is " + self.color)
class Dog(Animal):
     pass
dog = Dog(" Max", "4", "brown") 
 
dog.print_features()
 
class Cat(Animal):
      pass
cat = Cat( "Lucy", 4,"White")
 
cat.print_features()
```

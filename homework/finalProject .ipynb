```
class Company:
  def __init__(self, firstname, lastname,age, language ):
     self.name1 = firstname
     self.name2 = lastname
     self.age = age
     self.lang = language
 
  def fullDetails(self):
      print(self.name1 + " " + self.name2 +" Age: " + str (self.age) +" , " +  self.lang)
 
class Manager(Company):
     def __init__(self, firstname, lastname,age,language,level): 
          super().__init__(firstname,lastname,age,language)
          self.level = level
     
     def level1(self):
         print("I am a " + self.level)
     
 
Manager1 = Manager("Ahmed ", "ola", 28,"English", "Manager")
 
Manager1.fullDetails()
Manager1.level1()
```
